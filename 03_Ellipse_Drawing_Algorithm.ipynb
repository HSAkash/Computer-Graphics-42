# Ellipse Drawing Algorithm
* Mid Point Drawing Algorithm

## 1. Mid Point Drawing Algorithm

### 1.1. Import dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1.2. Variables initialization

```python
factor = 1
h = 0 * factor
k = 0 * factor
rx = 20 * factor
ry = 10 * factor
```

### 1.3. Algorithm
* Step 1: Find 1/4 1st region of the ellipse x and y coordinates
* Step 2: Find 1/4 2nd region of the ellipse x and y coordinates
* Step 3: Find 3/4 of the ellipse using 1/4 x and y coordinates
* Step 4: Add h and k values with x and y coordinates
* Step 5: Ellipse calculate with factor
* Step 6: Plot the Ellipse

#### 1.3.1. Step 1: Find 1/4 1st region of the ellipse x and y coordinates

```python
# Step 1: set X0, Y0 
x = [0]
y = [ry]

# Step 2: Finding discision parameter p = 1 - R
p = ry**2 - rx**2 * ry + 0.25 * rx**2

# Step 3: finding 1/8 circle potions
while 2*ry**2*x[-1] < 2*rx**2*y[-1]:
    x.append(x[-1] + 1)
    if p < 0:
        y.append(y[-1])
        p = p + 2 * ry**2 * x[-1] + ry**2
    else:
        y.append(y[-1] - 1)
        p = p + 2 * ry**2 * x[-1] - 2 * rx**2 * y[-1] + ry**2
```

#### 1.3.2. Step 2: Find 1/4 2nd region of the ellipse x and y coordinates

```python
while y[-1] > 0:
    y.append(y[-1] - 1)
    if p > 0:
        x.append(x[-1])
        p = p - 2 * rx**2 * y[-1] + rx**2
    else:
        x.append(x[-1]+1)
        p = p + 2 * ry**2 * x[-1] - 2 * rx**2 * y[-1] + rx**2
```

```python
x = np.array(x)
y = np.array(y)
```

#### 1.3.3. Step 3: Find 3/4 of the ellipse using 1/4 x and y coordinates

```python
ellipes = []
ellipes.append(np.array([x, y]).T)
ellipes.append(np.array([x, -y]).T)
ellipes.append(np.array([-x, -y]).T)
ellipes.append(np.array([-x, y]).T)
ellipes = np.array(ellipes)
```

#### 1.3.4. Step 4: Add h and k values with x and y coordinates

```python
# add h and k 
ellipes = ellipes + np.array([h, k])
```

#### 1.3.5. Step 5: Ellipse calculate with factor

```python
factor = 200
f_h = 0 * factor
f_k = 0 * factor
f_rx = 5 * factor
f_ry = 3 * factor

# Step 1: set X0, Y0 
x = [0]
y = [f_ry]

""" Region 1 """
# Step 2: Finding discision parameter p = 1 - R
p = f_ry**2 - f_rx**2 * f_ry + 0.25 * f_rx**2

# Step 3: finding 1/8 circle potions
while 2*f_ry**2*x[-1] < 2*f_rx**2*y[-1]:
    x.append(x[-1] + 1)
    if p < 0:
        y.append(y[-1])
        p = p + 2 * f_ry**2 * x[-1] + f_ry**2
    else:
        y.append(y[-1] - 1)
        p = p + 2 * f_ry**2 * x[-1] - 2 * f_rx**2 * y[-1] + f_ry**2

""" Region 2 """
while y[-1] > 0:
    y.append(y[-1] - 1)
    if p > 0:
        x.append(x[-1])
        p = p - 2 * f_rx**2 * y[-1] + f_rx**2
    else:
        x.append(x[-1]+1)
        p = p + 2 * f_ry**2 * x[-1] - 2 * f_rx**2 * y[-1] + f_rx**2



x = np.array(x)
y = np.array(y)

f_ellipes = []
f_ellipes.append(np.array([x, y]).T)
f_ellipes.append(np.array([x, -y]).T)
f_ellipes.append(np.array([-x, -y]).T)
f_ellipes.append(np.array([-x, y]).T)
f_ellipes = np.array(f_ellipes)


# add h and k 
f_ellipes = f_ellipes + np.array([f_h, f_k])
```

#### 1.3.6. Step 6: Plot the Ellipse

```python
# initialize the figure
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(15,5))

# plot the Normal distance between points
for i in range(len(ellipes)):
    ax[0].plot(ellipes[i][:,0], ellipes[i][:,1])
ax[0].plot(h,k,'ro')
ax[0].set(
    title = 'Mid point ellipes',
)


# coordinate multiplication factor
xlabels = ax[0].get_xticks()[1:-1]
ylabels = ax[0].get_yticks()[1:-1]
for i in range(len(f_ellipes)):
    ax[1].plot(f_ellipes[i][:,0], f_ellipes[i][:,1])
ax[1].plot(f_h,f_k,'ro')

ax[1].set(
    title = f'Mid Point ellipes (Factor {factor})',
    xticks = np.linspace(f_h - f_rx, f_h + f_rx, xlabels.shape[0]),
    yticks = np.linspace(f_k - f_ry, f_k + f_ry,  ylabels.shape[0]),
    xticklabels = xlabels,
    yticklabels = ylabels
    
)
plt.show()
```
